```python
# Multi D List
# --------------
# Multiple Row and Multiple columns

multi_a = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
multi_a
```

```python
'''
Nested For
---------------
Outer For loop -> Loop row
Inner For loop -> Loop column
'''
```

```python
for row in multi_a: # loop row
    for column in row: # loop column
        print(column)
    else:
        print('Inner Loop')
else:
    print('Outer Loop')
```

```python
# prices: prices = [100, 250, 75, 400, 120, 310].
# 1. Write a program that creates two new empty lists: cheap_items and 
# expensive_items.
# 2. Loop through the prices list. If a price is greater than 200, append it to 
# expensive_items. Otherwise, append it to cheap_items.
# 3. Print both lists and calculate the total count of items in each using the len() 
# function.
```

```python
prices = [100, 250, 75, 400, 120, 310]

cheap_items = []
expensive_items = []

for price in prices:
    if price > 200:
        expensive_items.append(price)
    else:
        cheap_items.append(price)

print(f'Cheap Items: {cheap_items}')
print(f'Expensive Items: {expensive_items}')
print(f'Total values in cheap items: {len(cheap_items)}')
print(f'Total values in expensive items: {len(expensive_items)}\n')
```

```python
# ) Create a dictionary 
# named gradebook: gradebook = {"Sam": [80, 85, 90], "Tara": [92, 88, 95]}
# 1. Add a new student named "Leo" with scores [70, 75, 80].
# 2. Update Sam's second score (index 1) from 85 to 88.
# 3. Delete the last score in Tara's list using the pop() method.
# 4. Use a for loop and .items() to print each student's name and their highest score 
# using the max() function
```

```python
gradebook = {'Sam': [80, 85, 90], "Tara": [92, 88, 95]}
```

```python
gradebook['Leo'] = [70, 75, 80]
```

```python
gradebook['Sam'][1] = 88
```

```python
gradebook['Tara'].pop()
```

```python
for idx, val in gradebook.items():
    print(f'{idx}\nHighest Marks: {max(val)}\n')
```
