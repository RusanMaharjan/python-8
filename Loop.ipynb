```python
# Write a program to print 10 natural numbers.
for i in range(1, 11, 1): # value - 1
    print(i)
```

```python
# Find odd and even using for loop for 20 numbers.
for i in range(1, 21):
    if i % 2 == 0:
        print(f'{i}: Even')
    else:
        print(f'{i}: Odd')
```

```python
# Multiplication Table
# 3 x 1 = 3
# 3 x 2 = 6
# ...
# 3 x 10 = 30
num = int(input('Enter a number: '))
stop_val = int(input('Enter stop number: '))

for i in range(1, stop_val+1):
    mul = num * i
    print(f'{num} x {i} = {mul}')
```

### Break and Continue

```python
for i in range(1, 11, 1): # value - 1
    if i == 5:
        break # stop program forcefully
    print(i)
```

```python
for i in range(1, 11, 1): # value - 1
    if i == 5:
        continue # skips value
    print(i)
```

```python
# Food Order System
print('Food Order System')
print('1. Breakfast')
print('2. Lunch')
print('3. Snaks')
print('4. Dinner')
print('5. Drinks')
print('6. Exit')

for i in range(1, 6):
    choose = int(input('Enter a number (1 - 6)'))
    
    if choose == 1:
        print('You have ordered Breakfast.')
    elif choose == 2:
        print('You have ordered Lunch.')
    elif choose == 3:
        print('You have ordered Snaks.')
    elif choose == 4:
        print('You have ordered Dinner.')
    elif choose == 5:
        print('You have ordered Drinks.')
    elif choose == 6:
        print('Thank you for choosing us.')
        break
    else:
        print('Invalid Number.')
else:
    print('Program Stopped. Order Limit!!')
```

```python
##### Match Case
print('Food Order System')
print('1. Breakfast')
print('2. Lunch')
print('3. Snaks')
print('4. Dinner')
print('5. Drinks')
print('6. Exit')

for i in range(1, 6):
    choose = int(input('Enter a number (1 - 6)')) # 1
    
    match choose:
        case 1:
            print('You have ordered Breakfast.')
        case 2:
            print('You have ordered Lunch.')
        case 3:
            print('You have ordered Snaks.')
        case 4:
            print('You have ordered Dinner.')
        case 5:
            print('You have ordered Drinks.')
        case 6:
            print('Thank You for choosing us.')
            break
        case _:
            print('Invalid number.')
else:
    print('Program ended. Order Limit.')
```

```python
# user can input pin 3 times. if pin is not correct.
# pin correct then user can extract balance.

# Card Blocked. Kindly visit nearest branch.
```

```python
pin = 1234
balance = 50000

for i in range(1, 4):
    user_pin = int(input('Enter your pin number: '))
    
    if user_pin == pin:
        user_balance = int(input('Enter your withdraw balance: '))
        if user_balance < balance:
            if user_balance % 1000 == 0:
                print(f'Rs.{user_balance} withdrawn from A/C 12345678.')
                rem_bal = balance - user_balance
                print(f'Remaining Balance: Rs.{rem_bal}')
                break
            else:
                print('Amount must be multiple of 1000.')
        else:
            print('Amount must be less than actual balance amount.')
    else:
        print('Pin number incorrect. Try again.')
else:
    print('Card Blocked. Kindly visit nearest branch.')
```

## While Loop

```python
# Guess the Number game 
# Write a program using while loop where: 
# Any number is defined at first: eg., secret_value = 7 
# The user keeps guessing until the correct number is guessed.
```

```python
secret_value = 7
guess_number = 0
attempts = 0

while guess_number != secret_value: # 7 != 7 False
    attempts += 1 # attempts = attempts + 1
    guess_number = int(input('Guess number from 1 - 1000: '))

    if guess_number == secret_value:
        # print('Total Attempts: ',attempts)
        print(f'Total Attempts: {attempts}')
        print('Correct Guess!!')
    else:
        print('Incorrect Guess. Try Again!!\n')

    if attempts == 3:
        print('No chance left. Thank You.')
        break
```
