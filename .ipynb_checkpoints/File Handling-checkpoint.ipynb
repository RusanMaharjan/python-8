## Text File

### Create text file

```python
# open(file_name, operation)
# .txt
user_input = input('Enter text: ')
file = open('test.txt', 'w') # Create File
content = file.write(user_input) # Add content to file
file.close() # Close File
```

```python
with open('new_text.txt', 'w') as file: # Create File
    file.write('New file content!!') # Add Content
```

### Read Text File

```python
file = open('test.txt', 'r')
content = file.read()
print(content)
```

```python
with open('new_text.txt', 'r') as file:
    content = file.read()
    print(content)
```

### Add Content in Existing Text File

```python
append_input = input('Enter text to append: ')
file = open('test.txt', 'a')
content = file.write(f'\n\t{append_input}')
file.close()
```

## CSV (Comma Separated Value) File

```python
import csv
```

```python
countries = [
    ['Country', 'Destination', 'Price'],
    ['Canada', 'Ottawa', 'Rs.8000'],
    ['Italy', 'Venice', 'Rs.4000'],
    ['Nepal', 'Pokhara', 'Rs.3000']
]
```

```python
# open(file_name, operation, newline='')
with open('country.csv', 'w', newline='') as file: # File Configuration
    write_file = csv.writer(file) # convert file to CSV File
    write_file.writerows(countries) # Insert Data to CSV File
```

### Read CSV File

```python
with open('country.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
new_countries = [
    ['India', 'New Delhi', 'Rs.8000'],
    ['Austrailia', 'Sydney', 'Rs.4000'],
    ['China', 'Hong Kong', 'Rs.3000']
]
```

```python
with open('country.csv', 'a', newline='') as file:
    append = csv.writer(file)
    append.writerows(new_countries)
```

```python
personal_details = [
    {
        'Name': 'Bob',
        'Age':30,
        'Address': 'Kathmandu',
        'Qualification': 'Bachelors'
    },
    {
        'Name': 'Charles',
        'Age':40,
        'Address': 'Lalitpur',
        'Qualification': 'Masters'
    },
    {
        'Name': 'Bill',
        'Age':20,
        'Address': 'Bhaktapur',
        'Qualification': 'High School'
    }
]
```

```python
with open('pd.csv', 'w', newline='') as file: # file configuration
    columns = ['Name', 'Age', 'Address', 'Qualification'] # column names
    writer = csv.DictWriter(file, fieldnames=columns) # create file ready for dictionary data and set column names
    writer.writeheader() # Write column names in csv file
    writer.writerows(personal_details) # insert data below header
```

```python
with open('pd.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        for key, val in row.items():
            print(f'{key}: {val}')
        print('-'*15)
```

```python
new_personal_details = [
    {
        'Name': 'Miles',
        'Age':30,
        'Address': 'Kathmandu',
        'Qualification': 'Bachelors'
    }
]
```

```python
with open('pd.csv', 'a', newline='') as file:
    columns = ['Name', 'Age', 'Address', 'Qualification']
    writer = csv.DictWriter(file, fieldnames=columns)
    writer.writerows(new_personal_details)
```

## JSON

```python
'''
write -> dump
read -> load
'''
```

```python
import json
```

```python
personal_details_json = [
    {
        'Name': 'Bob',
        'Age':30,
        'Address': 'Kathmandu',
        'Qualification': 'Bachelors'
    },
    {
        'Name': 'Charles',
        'Age':40,
        'Address': 'Lalitpur',
        'Qualification': 'Masters'
    },
    {
        'Name': 'Bill',
        'Age':20,
        'Address': 'Bhaktapur',
        'Qualification': 'High School'
    }
]
```

```python
with open('pd.json', 'w') as file:
    json.dump(personal_details_json, file, indent=4)
```

```python
with open('pd.json', 'r') as file:
    data = json.load(file)
    for row in data:
        for key, val in row.items():
            print(f'{key}: {val}')
        print('-'*15)
```

```python
new_personal_details_json = {
        'Name': 'Miles',
        'Age':30,
        'Address': 'Kathmandu',
        'Qualification': 'Bachelors'
    }

data.append(new_personal_details_json)

with open('pd.json', 'w') as file:
    json.dump(data, file, indent=4)
```
