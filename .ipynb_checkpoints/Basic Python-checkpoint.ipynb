# Python Operators

```python
'''
Operators
----------------
1. Arithmetic Operators
    -> +, -, *, /, % (remainder/modulus), //(float division)

2. Logical Operators
    -> AND, OR, NOT

3. Comparison Operator
    -> =, ==, <, >, <=, >=, !=
'''
```

```python
# input function
# input()
first_name = input('Enter your first name: ')
age = int(input('Enter your age: '))
print(first_name)
print(age)
```

```python
'''
Variables
-------------
1. Camel Case -> firstName, lastName, distanceToWork
2. Snake Case -> first_name, last_name, distance_to_work
'''
```

```python
num1 = int(input('Enter first number: '))
num2 = int(input('Enter second number: '))

add = num1 + num2
sub = num1 - num2
mul = num1 * num2
div = num1 / num2
rem = num1 % num2
fd = num1 // num2
```

```python
# \n -> new line \t -> tab
print(f'Addition: {add}\nSubtraction: {sub}')
# print('Addition:'+add+'\n'+'Subtraction:'+sub)
print(mul)
print(div)
print(rem)
print(fd)
```

```python
num1 = 5
num2 = 10
num3 = 6
```

```python
(num1 < num2) != (num2 < num3)
```

```python
# 1-> True / 0 -> False
# AND (A.B)
# -----
# T T -> T
# T F -> F
# F T -> F
# F F -> F

# OR (A+B)
# --------
# T T -> T
# T F -> T
# F T -> T
# F F -> F

# NOT
# --------
# T -> F
# F -> T
```
