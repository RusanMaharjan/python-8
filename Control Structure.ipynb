### Student Grade Calcuation

```python
# marks more than 100 -> Invalid marks. Must be between 0 and 100.
# input of 5 subject. Total Marks of all subject. Avergae Marks (Percentage)

sub1 = float(input('Enter your marks for Subject 1: '))
sub2 = float(input('Enter your marks for Subject 2: '))
sub3 = float(input('Enter your marks for Subject 3: '))
sub4 = float(input('Enter your marks for Subject 4: '))
sub5 = float(input('Enter your marks for Subject 5: '))
name_of_student = 'Ram'
total_marks = sub1 + sub2 + sub3 + sub4 + sub5
# format_String
print(f'Name: {name_of_student}. Total Marks of 4 Subjects: {total_marks}')
# print('Name: '+name_of_student+'Total Marks of 4 subjects:'+total_marks)

marks = total_marks/5
print(f'Average Marks of Student: {marks}')

if marks >= 80 and marks <= 100:
    print('Grade A')
elif marks >= 60 and marks < 80:
    print('Grade B')
elif marks >= 50 and marks < 60:
    print('Grade C')
elif marks >= 40 and marks < 50:
    print('Grade D')
elif marks >= 0 and marks < 40:
    print('Fail')
else:
    print('Invalid marks. Must be between 0 and 100.')
```

## ATM

```python
# 1. Pin Correct
# 2. withdraw amount less than balance amount
pin = 1234
balance = 50000

user_pin = int(input('Enter your pin number: '))
user_balance = int(input('Enter your withdraw balance: '))

if user_pin == pin and user_balance < balance:
    print(f'Rs.{user_balance} is withdrawn from A/C 12345678.')
    rb = balance - user_balance
    print(f'Remaining Balance: Rs.{rb}')
elif user_pin != pin and user_balance < balance:
    print('Incorrect Pin Number.')
elif user_pin == pin and user_balance > balance:
    print('Insufficient Balance.')
else:
    print('Incorrect Pin Number and Insufficient Balance.')
```

```python
'''
calculate tax based on annual income 
<=250000 then no tax 
250001 - 500000 then 5% tax 
500001 - 1000000 then 20% 
more than 1000000 then 30%

output:
Annual Income: 400000
Tax Amount: 20000
Remaining Salary: 380000
'''
```

```python
annual_income = int(input('Enter your annual income: '))

# <250000
if annual_income >=0 and annual_income <= 250000:
    print(f'Annual Income: Rs.{annual_income}')
    print(f'Tax Payable: Rs.0')
    remaining_salary = annual_income - 0
    print(f'Total Remaining Salary: Rs.{remaining_salary}')
# > 250000 & <= 500000
elif annual_income > 250000 and annual_income <= 500000:
    print(f'Annual Income: Rs.{annual_income}.')
    tax_payable = annual_income * 0.05
    print(f'Tax Payable(5%): Rs.{tax_payable}.')
    remaining_salary = annual_income - tax_payable
    print(f'Total Remaining Salary: Rs.{remaining_salary}')
# > 500000 & <= 1000000
elif annual_income > 500000 and annual_income <= 1000000:
    print(f'Annual Income: Rs.{annual_income}.')
    tax_payable = annual_income * 0.10
    print(f'Tax Payable(10%): Rs.{tax_payable}.')
    remaining_salary = annual_income - tax_payable
    print(f'Total Remaining Salary: Rs.{remaining_salary}')
# > 1000000
elif annual_income > 1000000:
    print(f'Annual Income: Rs.{annual_income}.')
    tax_payable = annual_income * 0.30
    print(f'Tax Payable(30%): Rs.{tax_payable}.')
    remaining_salary = annual_income - tax_payable
    print(f'Total Remaining Salary: Rs.{remaining_salary}')
else:
    print('Amount must be positive and more than 0.')
```

```python
# A student is eligible for an exam if: 
# Attendance >= 75 
# internal marks >= 40 
# print exact reason if not eligible
```

```python
attendance = int(input('Enter attendance %: '))
internal_marks = float(input('Enter marks: '))

if attendance >= 75 and internal_marks >= 40:
    print('You are eligible for examination.')
elif attendance <= 75 and internal_marks >= 40:
    print('Not eligible for exam. Attendance Requirement does not meet.')
elif attendance >= 75 and internal_marks <= 40:
    print('Not eligible for exam. Internal Marks Requirement does not meet.')
else:
    print('You are not eligible for exam.')
```

```python
pin = 1234
balance = 50000

user_pin = int(input('Enter your pin number: '))

if user_pin == pin:
    user_balance = int(input('Enter your withdraw balance: '))
    if user_balance < balance:
        if user_balance % 1000 == 0:
            print(f'Rs.{user_balance} withdrawn from A/C 12345678.')
            rem_bal = balance - user_balance
            print(f'Remaining Balance: Rs.{rem_bal}')
        else:
            print('Amount must be multiple of 1000.')
    else:
        print('Amount must be less than actual balance amount.')
else:
    print('Pin number incorrect. Try again.')
```

```python
# match case
number = int(input('Enter a number (1 - 7): '))

match number: # 2
    case 1:
        print('Sunday')
    case 2:
        print('Monday')
    case 3:
        print('Tuesday')
    case 4:
        print('Wednesday')
    case 5:
        print('Thursday')
    case 6:
        print('Friday')
    case 7:
        print('Saturday')
    case _: # default
        print('Invalid Number')
```

```python
print('Restaurant Menu')
print('*'*30)
print('1. Momo')
print('2. Chowmin')
print('3. Cold Drinks')
print('4. Hot Drinks')
print('5. Exit\n')
attempts = 5
for i in range(1, 6):
    remaining_attempts = attempts - i
    print(f'Remaining Orders: {remaining_attempts}')
    print('-'*30)
    choose = input('Enter order number from 1 - 5: ')
    
    match choose:
        case '1':
            print('You have ordered Momo.\n')
        case '2':
            print('You have ordered Chowmin.\n')
        case '3':
            print('You have ordered Cold Drinks.\n')
        case '4':
            print('You have ordered Hot Drinks.\n')
        case '5':
            print('Thank You for choosing us!')
            break
        case _:
            print('Invalid input: Choose order value from 1 to 5.\n')
```
