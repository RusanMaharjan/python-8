```python
# one (Parent Class) to many (Child Class)
```

```python
# Customer <-> Order
'''
- Child class inherits/extends parent class (variables, functions, details)
'''
```

```python
# Parent Class
class Customers:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone):
        self.customer_id = customer_id
        self.customer_name = customer_name
        self.customer_address = customer_address
        self.customer_phone = customer_phone

    def customer_details(self):
        print(f'\nCustomer Details')
        print('-'*50)
        print(f'Customer ID: {self.customer_id}')
        print(f'Customer Name: {self.customer_name}')
        print(f'Customer Address: {self.customer_address}')
        print(f'Customer Phone: {self.customer_phone}')
```

```python
class Orders(Customers):
    def __init__(self, order_id, order_name, order_desc, order_destination, order_price,
                customer_id, customer_name, customer_address, customer_phone):
        super().__init__(customer_id, customer_name, customer_address, customer_phone)
        self.order_id = order_id
        self.order_name = order_name
        self.order_desc = order_desc
        self.order_destination = order_destination
        self.order_price = order_price

    def order_details(self):
        print(f'\nOrder Details')
        print('-'*50)
        print(f'Order Id: {self.order_id}')
        print(f'Order Name: {self.order_name}')
        print(f'Order Description: {self.order_desc}')
        print(f'Order Destination: {self.order_destination}')
        print(f'Order Price: {self.order_price}')
```

```python
o = Orders(1, 'Mobile Order', 'Samsung Mobile', 'Bhaktapur', 30000, 12, 'Bob', 'Kathmandu', 1234567890)
```

```python
o.order_details()
o.customer_details()
```

```python
SCHOOL_NAME = 'DEF School'
class School:
    def __init__(self, school_address, school_contact):
        self.school_name = SCHOOL_NAME
        self.school_address = school_address
        self.school_contact = school_contact

    def school_details(self):
        print(f'School Name: {self.school_name}')
        print(f'School Address: {self.school_address}')
        print(f'School Contact: {self.school_contact}')
```

```python
class Teacher(School):
    def __init__(self, teacher_id, teacher_name, teacher_email, teacher_phone, no_of_subject, teacher_qualification,
                school_address, school_contact):
        super().__init__(school_address, school_contact)
        self.teacher_id = teacher_id
        self.teacher_name = teacher_name
        self.teacher_email = teacher_email
        self.teacher_phone = teacher_phone
        self.no_of_subject = no_of_subject
        self.teacher_qualification = teacher_qualification

    def teacher_details(self):
        print(f'Teacher Id: {self.teacher_id}')
        print(f'Teacher Name: {self.teacher_name}')
        print(f'Teaher Email: {self.teacher_email}')
        print(f'Teacher Phone: {self.teacher_phone}')
        print(f'Number of Subject: {self.no_of_subject}')
        print(f'Teacher Qualificaiton: {self.teacher_qualification}')
```

```python
class Student(School):
    def __init__(self, student_id, student_name, student_email, student_phone, student_section,
                school_address, school_contact):
        super().__init__(school_address, school_contact)
        self.student_id = student_id
        self.student_name = student_name
        self.student_email = student_email
        self.student_phone = student_phone
        self.student_section = student_section

    def student_details(self):
        print(f'Student Id: {self.student_id}')
        print(f'Student Name: {self.student_name}')
        print(f'Student Email: {self.student_email}')
        print(f'Student Phone: {self.student_phone}')
        print(f'Student Section: {self.student_section}')
```

```python
t = Teacher(1, 'Bob', 'bob@gmail.com', '1234567890', 7, 'Bachelors', 'Kathmandu', '01-1234567')
t.teacher_details()
t.school_details()
```

```python
s = Student(101, 'Hari', 'hari@gmail.com', 9876541230, 'C11', 'Kathmandu', '01-5432167')
s.student_details()
s.school_details()
```

```python
class Teacher(School):
    def __init__(self, teacher_id, teacher_name, teacher_email, teacher_phone, no_of_subject, teacher_qualification,
                school_address, school_contact):
        super().__init__(school_address, school_contact)
        self.teacher_id = teacher_id
        self.teacher_name = teacher_name
        self.teacher_email = teacher_email
        self.teacher_phone = teacher_phone
        self.no_of_subject = no_of_subject
        self.teacher_qualification = teacher_qualification
```

## Encapsulation

```python
'''
Encapsulation is used for data hiding.
- Access Modifiers
---------------------
1. Private (self.__variable_name = variable_name)
2. Public (self.variable_name = variable_name)
3. Protected (self._variable_name = variable_name)
'''
```

### Private Access Modifiers

```python
class Brand:
    def __init__(self, brand_id, brand_name):
        self.__brand_id = brand_id
        self.__brand_name = brand_name

    def brand_details(self):
        print(f'Brand Id: {self.__brand_id}')
        print(f'Brand Name: {self.__brand_name}')
```

```python
class Product(Brand):
    def __init__(self, product_id, product_name, product_price, brand_id, brand_name):
        super().__init__(brand_id, brand_name)
        self.__product_id = product_id
        self.__product_name = product_name
        self.__product_price = product_price

    def product_details(self):
        print(f'Product Id: {self.__product_id}')
        print(f'Product Name: {self.__product_name}')
        print(f'Product Price: {self.__product_price}')
```

```python
b = Brand(1, 'Samsung')
b.brand_details()
```

```python
p = Product(101, "Washing Machine", 100000, 1, 'Samsung')
p.product_details()
p.brand_details()
```

```python
'''
Multi-Level Inheritance -> Encausulation (Private)
Relation: Customers order handeled by staff
------------------------------------------------
Customer (cust_id, cust_name, cust_add, cust_contact, cust_email)
Staff (staff_id, staff_name, staff_add, staff_contact, staff_email)
Orders (order_id, order_name, order_desc, order_destination, order_price)

order -> Object
'''
```

```python
class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_contact, customer_email):
        self.__customer_id = customer_id
        self.__customer_name = customer_name
        self.__customer_address = customer_address
        self.__customer_contact = customer_contact
        self.__customer_email = customer_email

    def customer_details(self):
        print(f'Customer ID: {self.__customer_id}')
        print(f'Customer Name: {self.__customer_name}')
        print(f'Customer Address: {self.__customer_address}')
        print(f'Customer Contact: {self.__customer_contact}')
        print(f'Customer Email: {self.__customer_email}')
```

```python
class Staff(Customer):
    def __init__(self, staff_id, staff_name, staff_address, staff_contact, staff_email,
                customer_id, customer_name, customer_address, customer_contact, customer_email):
        super().__init__(customer_id, customer_name, customer_address, customer_contact, customer_email)
        self.__staff_id = staff_id
        self.__staff_name = staff_name
        self.__staff_address = staff_address
        self.__staff_contact = staff_contact
        self.__staff_email = staff_email

    def staff_details(self):
        print(f'Staff ID: {self.__staff_id}')
        print(f'Staff Name: {self.__staff_name}')
        print(f'Staff Address: {self.__staff_address}')
        print(f'Staff Contact: {self.__staff_contact}')
        print(f'Staff Email: {self.__staff_email}')
```

```python
class Orders(Staff):
    def __init__(self, order_id, order_name, order_description, order_destination, order_price,
                staff_id, staff_name, staff_address, staff_contact, staff_email,
                customer_id, customer_name, customer_address, customer_contact, customer_email):
        super().__init__(staff_id, staff_name, staff_address, staff_contact, staff_email,
                customer_id, customer_name, customer_address, customer_contact, customer_email)
        self.__order_id = order_id
        self.__order_name = order_name
        self.__order_description = order_description
        self.__order_destination = order_destination
        self.__order_price = order_price

    def order_details(self):
        print(f'Order Id: {self.__order_id}')
        print(f'Order Name: {self.__order_name}')
        print(f'Order Description: {self.__order_description}')
        print(f'Order Destination: {self.__order_destination}')
        print(f'Order Price: {self.__order_price}')
```

## Polymorphism

```python
'''
Poly -> Many, Morph -> Form
- One function multiple work.
'''
```

```python
from datetime import datetime
```

```python
class Viber:
    def message(self, send_from, send_to, message):
        print(f'Message Send From Viber.\nFrom: {send_from}\nTo: {send_to}\nMessage: {message}')
        print(f'Time: {datetime.now()}')

class WhatsApp:
    def message(self, send_from, send_to, message):
        print(f'Message Send From WhatsApp.\nFrom: {send_from}\nTo: {send_to}\nMessage: {message}')
        print(f'Time: {datetime.now()}')

class Messenger:
    def message(self, send_from, send_to, message):
        print(f'Message Send From Messenger.\nFrom: {send_from}\nTo: {send_to}\nMessage: {message}')
        print(f'Time: {datetime.now()}')

class Slack:
    def message(self, send_from, send_to, message):
        print(f'Message Send From Slack.\nFrom: {send_from}\nTo: {send_to}\nMessage: {message}')
        print(f'Time: {datetime.now()}')
```

```python
def send_message(message_obj, send_from, send_to, message):
    message_obj.message(send_from, send_to, message)
```

```python
send_message(Slack(), 'Ram', 'Shyam', 'Hello.')
```

## Abstracton

```python
'''
-- Interface -> Class without body
-- Abstract Class
-- Atleast one function must be an abstract function
'''
```

```python
# Abstract Base Class
from abc import ABC, abstractmethod
```

```python
class Bank(ABC):
    # annotation / decorator
    @abstractmethod
    def accountDetails(self):
        pass

    @abstractmethod
    def statement(self):
        pass
    
    def userDetails(self):
        print(f'Name: Ram')
        print(f'Address: Kathmandu')
```

```python
class MobileBanking(Bank):
    def connection(self):
        return 'Connected to server!!'

    def accountDetails(self):
        print(f'Your account number: 12345678909876543.')
        print(f'Your balance amount: Rs.50000')

    def statement(self):
        print('Statement from date1 to date 2...')
```

```python
mb = MobileBanking()
```

```python
mb.accountDetails()
```

```python
mb.statement()
```

```python
class Value:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def val(self):
        return Value(self.a < self.b)
```
