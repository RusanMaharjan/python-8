# Error Handling

```python
'''
try, except(catch), else, finally, custom exception
'''
```

```python
try:
    fkdsjakfjld
except:
    print('Error')
```

```python
try:
    num1 = int(input('Enter first number: '))
    num2 = int(input('Enter second number: '))
    
    num3 = num1 / num2
    print(num3)
except:
    print('Error Found!!')
```

```python
try:
    num1 = int(input('Enter first number: '))
    num2 = int(input('Enter second number: '))
    
    num3 = num1 / num2
    print(num3)
except ZeroDivisionError:
    print('Zero Division Error: Number cannot be divisible by 0.')
except ValueError:
    print('Value Error: Only number integer can be pass as an argument in input field.')
```

```python
try:
    file_name = input('Enter your file name: ')
    extension = input('Enter your file extention: ')
    with open(f'{file_name}.{extension}', 'r') as file:
        content = file.read()
except FileNotFoundError as e:
    print(e)
else:
    print(content)
finally:
    print(f'Program ended successfully.')
```

```python
'''
Voting System
---------------
age >= 18
'''
```

```python
class AgeError(Exception):
    pass
```

```python
def checkAge(age):
    if age >= 18:
        print('You are eligible to vote.')
    else:
        raise AgeError(f'{age} age is not eligible to vote.')
```

```python
try:
    age = int(input('Enter your age: '))
    checkAge(age)
except AgeError as e:
    print(e)
except ValueError:
    print('Value Error: Only number integer can be sent to age input field.')
```
