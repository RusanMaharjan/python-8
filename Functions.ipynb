## default argument and keyword argument

```python
def employee_details(name, age=0, gender='M', jobPosition='Unknown Position', 
                     jobType='Unknown Type'):
    print(f'Name of Employee: {name}')
    print(f'Age of Employee: {age}')
    print(f'Gender of Employee: {gender}')
    print(f'Job Position of Employee: {jobPosition}')
    print(f'Job Type of Employee: {jobType}')
```

```python
employee_details('Ram', 30, 'M')
```

```python
employee_details('Ram', 'M', jobType='Part-Time', jobPosition='Data Analyst')
```

## All Arguments (*args) and All Keyword Arguments (**kwargs)

```python
def employeeDetails(*details):
    print(details)
    for data in details:
        print(data)
```

```python
employeeDetails('Ram', 30, 'Kathmandu', 'Bachelors', 'Full-Time', 'Data Scientist')
```

```python
def employeeDetailsKwargs(**kwargs):
    print(kwargs)
    for key, val in kwargs.items():
        print(f'{key} of Employees: {val}')
```

```python
employeeDetailsKwargs(Name='Ram', Age=30, JobType='Full-Time', JobPosition='Data Analyst')
```

## Nested Function

```python
def add(x, y):
    print(f'Addition of {x} and {y} is: {x + y}')

def sub(x, y):
    print(f'Subtraction of {x} and {y} is: {x - y}')

def mul(x, y):
    print(f'Multiplication of {x} and {y} is: {x * y}')

def div(x, y):
    print(f'Division of {x} and {y} is: {x / y}')
```

```python
def calculator(x, y):
    add(x, y)
    sub(x, y)
    mul(x, y)
    div(x, y)
```

```python
calculator(10, 5)
```

```python
def add(x, y):
    return f'Addition of {x} and {y} is: {x + y}'

def sub(x, y):
    return f'Subtraction of {x} and {y} is: {x - y}'

def mul(x, y):
    return f'Multiplication of {x} and {y} is: {x * y}'

def div(x, y):
    return f'Division of {x} and {y} is: {x / y}'
```

```python
def calculator(x, y):
    return add(x, y), sub(x, y), mul(x, y), div(x, y)
```

```python
data = calculator(10, 5)
for d in data:
    print(d)
```

## Lambda Function

```python
# One Line Function
# function_name = lambda arg1, arg2 : logic / condition / print...
```

```python
def oddEven(num):
    if num % 2 == 0:
        print('Even')
    else:
        print('Odd')
oddEven(6)
```

```python
add = lambda num1, num2 : num1 + num2
add(5, 10)
```

```python
odd_even = lambda num : "Even" if num % 2 == 0 else "Odd"
odd_even(1)
```
